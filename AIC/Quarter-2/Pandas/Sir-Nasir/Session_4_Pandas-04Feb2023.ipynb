# Merging and Joining  Data Frame

```python
import pandas as pd
import numpy as np
import datetime
customers  = {'customer_id':[1234,1235,1111,2222,3334,3333,4444,5555],
             'Name':['Mike', 'Maria', 'Nasir', 'Ali',"Ahmed", 'Saad',"Hamid", "Kami"],
             'Address':['abc23-1', 'xyz-3','abc23-2', 'xyz-5','abc23-6', 'xyz-0','abc23-11', 'xyz-301'],
             }

customers = pd.DataFrame(customers)
customers
```

```python
customers.info()
```

```python
# customers.set_index('customer_id')
```

```python
orders = {'customer_id':[1234,1235,1234,1234,1235,1267,1237,1890],
          'order_id':[11,22,33,44,55,66,77,88],
         'order_date':[datetime.date(2023,2,1),datetime.date(2023,1,1),
                       datetime.date(2023,2,2),datetime.date(2023,1,1),
                       datetime.date(2023,2,3),datetime.date(2023,1,2),
                       datetime.date(2023,2,4),datetime.date(2023,1,2)],
         }

orders = pd.DataFrame(orders)
orders
```

```python
orders.info()
```

# Merge
  - Inner 
  - Outer
  - Left 
  - Right

```python
merged_data = pd.merge(customers,orders)
# inner >> intersection
```

```python
merged_data.sort_values(by='customer_id',ascending=False)
```

```python
merged_data = pd.merge(orders,customers)
```

```python
merged_data
```

```python
merged_data = pd.merge(orders,customers,how='inner')
merged_data
```

```python
merged_data = pd.merge(customers,orders, how='outer')# outer merger >>> union
merged_data
```

```python
merged_data = pd.merge(customers,orders, how='left')
merged_data
```

```python
merged_data = pd.merge(customers,orders, how='right')
merged_data
```

```python
df1 = pd.DataFrame({'left': ['foo', 'bar']})
df2 = pd.DataFrame({'right': [7, 8]})
df = pd.merge(df1,df2, how='cross')
df
```

```python
df1 = pd.DataFrame({'key':[1,1],'left': ['foo', 'bar']})
df2 = pd.DataFrame({'key':[1,1],'right': [7, 8]})
df = pd.merge(df1,df2, how='cross')[['left','right']]
df
```

```python
customers  = {'customerID':[1234,1235,1111,2222,3334,3333,4444,5555],
             'Name':['Mike', 'Maria', 'Nasir', 'Ali',"Ahmed", 'Saad',"Hamid", "Kami"],
             'Address':['abc23-1', 'xyz-3','abc23-2', 'xyz-5','abc23-6', 'xyz-0','abc23-11', 'xyz-301'],
              'order_date':[datetime.date(2023,2,1),datetime.date(2023,1,1),
                       datetime.date(2023,2,2),datetime.date(2023,1,1),
                       datetime.date(2023,2,3),datetime.date(2023,1,2),
                       datetime.date(2023,2,4),datetime.date(2023,1,2)]
             }

customers = pd.DataFrame(customers)

orders = {'customerID':[1234,1235,1234,1234,1235,1267,1237,1890],
          'order_id':[11,22,33,44,55,66,77,88],
         'order_date':[datetime.date(2023,2,1),datetime.date(2023,1,1),
                       datetime.date(2023,2,2),datetime.date(2023,1,1),
                       datetime.date(2023,2,3),datetime.date(2023,1,2),
                       datetime.date(2023,2,4),datetime.date(2023,1,2)],
         }

orders = pd.DataFrame(orders)
```

```python
customers
```

```python
orders
```

```python
df_merged = pd.merge(customers,orders)
df_merged
```

```python
df_merged = pd.merge(customers,orders, on='order_date',suffixes=("_left", "_right"))
df_merged
```

```python
df_merged = pd.merge(customers,orders, on=['order_date', 'customerID'],suffixes=("_left", "_right"))
df_merged
```
