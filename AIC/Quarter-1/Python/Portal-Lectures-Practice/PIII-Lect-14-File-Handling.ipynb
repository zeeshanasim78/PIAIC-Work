```python
# PIII Lecture 14 : File Handling 
# We can do the 
f=open("records.txt","w+")

# The above mentioned course will open the records.txt file if such a file exists. If this file does not exists Python wil;l create that file
# With is the way to tell Python to close the file after you write on it
# There are three modes in the File Handling
# Read , Write , Append

# To write in the file we use the write function
f.write("Hello, World")
# The above code tells Python to replace any data that is already in file . 

f.close()




```

```python
# a file named "record.txt", will be opened with the reading mode.
file = open("records.txt", "r")
# This will print every line one by one in the file
for each in file:
  print(each)
```

```python
## Open file using with statement
with open("records.txt", "a") as my_file:
    my_file.write("Hello world \n")
    my_file.write("I hope you're doing well today \n")
    my_file.write("This is a text file \n")
    my_file.write("Have a nice time \n")

with open("records.txt") as my_file:
    print(my_file.read())

# Output: 
# Hello world 
# I hope you're doing well today
# This is a text file
# Have a nice time
```

```python
with open("hello.txt", "w") as my_file:
    my_file.write("Hello world \n")
    my_file.write("I hope you're doing well today \n")
    my_file.write("This is a text file \n")
    my_file.write("Have a nice time \n")

with open("hello.txt") as my_file:
    for line in my_file:
        print(line)

# Output:
# Hello world 

# I hope you're doing well today 

# This is a text file

# Have a nice time 
```
