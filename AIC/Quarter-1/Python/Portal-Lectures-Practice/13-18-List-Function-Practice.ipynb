```python
# Python Lists and its Functions
'''  
A List is a data structure in Python that is mutable (Cannot be changed) or changeable, ordered sequence of elements.
List elements are accessed by providing the identifier and index # element inside square brackets
'''
stds=[21001,"Zeeshan", "Asghar Khan", "Fsc", 876 , 75.35]
# Index# 0       1        2             3      4      5

print("Items in list stds are : ", stds)
# To print value at index # 2 and 5 we use following command
print(" Value at index 2 of list is = ", stds[2])
print(" Value at index 5 of list is = ", stds[5])

# To Add values in the List at the end we use append()
stds.append("Pre Engg")  
# Note : Only 1 Value can be appended at a time
print("\n Items in list after appending value : ", stds)

# To Add value at a specific index # we use the Insert fuction()
stds.insert(3," Fauji Foundation Model School")  
# Note : 1st Parameter  location index no to add  , 2nd is the in item to be addedd
print("\n Items in list after Inserting at Index 3 : ", stds)

# To add multiple values at the end of list we use extend()
# This function requires a IIterable object as parameter which can be an other list
stds.extend(["Math", "Physics","Computer"])
print("List after adding 3 values using extend()",stds)

# To count how many times a item appears in the list we use count()
print("The value Zeeshan exists following times in the list",stds.count("Zeeshan"))

# To find index of a item in the list we use index()
print("The value Zeeshan is present at index # ",stds.index("Zeeshan"))

# We can also use the in operator to check if a value exists inside a List
var="Zeeshan1"
if var in stds:
    print("The Value" , var , " exists  at index # ", stds.index(var))
else:
  print("The Value" , var , " does not exists in list stds ", stds)


#To clear all items permanently from the list we use clear()
stds.clear()

stds=[21001,"Zeeshan", "Asghar Khan", "Fsc", 876 , 75.35]
# Index# 0       1        2             3      4      5

# To create copy of a list we can use copy(). This function copies values by Value
list2_stds=stds.copy() 
# Copy by Value means a new variable list2_stds will be  created in the memory locations
# that have the same values as stds and change in stds list will not be visible in list2_std
list2_stds.append(" list2_std ")
stds.append(" original list stds")
print("\n The value of stds list = ", stds)
print("The value of list2_stds = ", list2_stds)

stds=[21001,"Zeeshan", "Asghar Khan", "Fsc", 876 , 75.35]
# Index# 0       1        2             3      4      5
# To use copy by Reference we use the assignment operator
list3_stds =stds  # This is copy by Reference
# Now  in this case change in one list will also be reflected in both lists
stds.append("New Append Value")
print("\n The value of stds list = ", stds)
print("The value of list3_stds = ", list3_stds)

# To remove item from a list we use del key word as
print("\n The value of list3_stds = ", list3_stds)
del list3_stds[3]
print(" The value of list3_stds after deletion of item at  = ", list3_stds)
del list3_stds
print(" The list3_stds deleted")


```

```python
# Slicing in Python Lists 
'''  
Python provides a very handy feature if you want to take a partial piece of list we use slicing option
Note: As slicing operation copies elements from the original list , hence this operation returns a new list
which is a subset of  original list
'''
cities=["Karachi", "Lahore", "Quetta", "Peshawar", "Multan", "Islamabad"]
# Index#     0         1        2             3         4        5
#Resverse Idx -6      -5       -4            -3        -2       -1
# listname[start index : end index : step]
# Note : Start Index is Included where as End Index no is not included
# e.g: cities[1:3] # By defualt step is 1
print("cities[1:3] will return ", cities[1:3] , " from original list ", cities)

print("cities[-2:-1] will return ", cities[-2:-1] , " from original list ", cities)

print("cities[1: :2] will return ", cities[1: : 2] , " from original list ", cities)

print("cities[-4:-1] will return ", cities[-4:-1] , " from original list ", cities)

print("cities[-5:-1:2] will return ", cities[-5:-1:2] , " from original list ", cities)
```

```python
# Slicing in Python Lists 
# Slicing using start and end values and Negative Indexing

students=["Ali", "Faisal", "Qamar", "Palwasha", "Muzafar", "Imran","Tanveer"]
# Idx #     0        1        2         3           4          5         6
# -ive Idx -7       -6       -5        -4          -3         -2        -1

# listname[start : end]  # in this statement start element is included but end is not included
# So in above list of students if we want to get values from index # 2 to three then we will use as
print(students[2:3], " are Items from List Students wher start index is 2 anad end index is 3")  
print(students[2:4], " are Items from List Students wher start index is 2 anad end index is 4 \n")  

# If we did not specify any start and end value then the Python will display all items in the list
print(students[:], " All items in list are printed as we did not specified start and end value \n")

# If we only specify start or end index then 
print(students[3:], " All items in list from index 3 will be printed as we did not specified end value \n")
print(students[:4], " All items in list from up to index 3 to  will be printed as we did not specified start value but given end value\n")

# Using Negative indexing
print("Nagative Indexing in Lists")
# In Python, you can start indexing from the end of an iterable. This is known as negative indexing.
# We can use both positive and negative indexes to access iterables / list.
# Let us consider and example of using postive and negative indexing at a time as
print("students[2:-1]")
print(students[2:-1], "\n It will get items from index 2 to index 4  or index -2  as last index not included \n")

print(" If we using list as students[2:-5]") 
print(students[2:-5]) # Empty List
print(" In python if the list index are not specified correctly it will not generate error but will return null list")
```
