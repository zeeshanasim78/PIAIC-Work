```python
# Steps for database manipultion
# First we will make the query. So we will now write the query and store it in a string variable
# In 2nd step we run query stored in string variable in Python with the open database connection and cursor to perform database manipuation
# In the 3rd step we will commit the tranaction so that the changes made in the datebase are updated in database

conn=None
cur=None

try:
    conn  = psycopg2.connect(host=host_name, dbname=data_base , user=user_name, password=pwd,port=port_id)  
    print("Connection Establised")
    
    # In order to do transaction with database we will need to open a cursor
    # A cursror is something that helps you performt he transaction
    
    # The statement will open the cursor
    cur=conn.cursor()
    
    # Create the query
    create_script = '''
                    create table if not exists dept ( 
                    id int primary key
                    , dept_name varchar(40) not null
                    , purpose varchar(300)
                    )
    '''
    
    # Now we will execute the string on the open cursor object as 
    cur.execute(create_script)
    print("Table Created")
    
    # We need to commit it to update changes in the database
    conn.commit()
    print("Changes Commited in the database")
        
   
except Exception as error:
    print("Exception : " + str(error))
else:
    print("No error generated during execution of Try Block")
finally:
    # It is a good practice to add cursor close and database close statement in the finally block so that these statements are executed always
    if cur is not None:
        cur.close()
        print("Cursor Closed")
        
    if conn is not None:
        conn.close()
        print("Connection Closed")
```

```python
# Now we will see how can we execute mutiple values using insert statement on the database from Python

conn=None
cur=None

try:
    conn  = psycopg2.connect(host=host_name, dbname=data_base , user=user_name, password=pwd,port=port_id)  
    print("Connection Establised")
    
    # In order to do transaction with database we will need to open a cursor
    # A cursror is something that helps you performt he transaction
    
    # The statement will open the cursor
    cur=conn.cursor()
    
       
    # Drop Table Script 
    drop_script = '''
            drop table dept
    '''
    # Execute the drop table script to delete table from database
    cur.execute(drop_script)
    print("Dept Table removed from database")
    
    # Now we will write create table query and store it in a string variable
    create_script = '''
                    create table if not exists dept ( 
                    id int primary key
                    , dept_name varchar(40) not null
                    , purpose varchar(300)
                    )
    '''
    #Excute the create table script
    cur.execute(create_script)
    print("New Dept Table created in the database")
    
    
    # Create the query with place holders
    insert_query_script = '''
                   insert into dept(id, dept_name, purpose)
                   values (%s,%s,%s)
    '''
    
    # define the data to insert in the table using place holder query
    insert_values = [(1,"HQ Div", " To manage all the resoruces & procurement for the HQ")
                     ,(2,"HRM Div", " To manage all the Human Resoruce Activities for the HQ")
                     ,(3,"AP Div", " To manage all the Awareness and PRevention Activities for the HQ")
                     ,(4,"Pros Div", " To manage all the Prosecution Activities & Court Matters of the HQ")
                    ]
    # This for loop will each tuple from the list and execute them one by one to add a record in the database
    for record in insert_values:
        cur.execute(insert_query_script,record)
        print(record, " values insert in the database")
        
        
    # We need to commit it to update changes in the database
    conn.commit()
    print("Changes Commited in the database")
        
    
except Exception as error:
    print("Exception : " + str(error))
else:
    print("No error generated during execution of Try Block")
finally:
    # It is a good practice to add cursor close and database close statement in the finally block so that these statements are executed always
    if cur is not None:
        cur.close()
        print("Cursor Closed")
        
    if conn is not None:
        conn.close()
        print("Connection Closed")
```

```python
# Now we will execute Select statement to retrieve data from the database usnig Python Code
# We need two packages for this purpose
import psycopg2  # For the database connectivity
import psycopg2.extras # To open the cursor for select and manage data


conn=None
cur=None

try:
    conn  = psycopg2.connect(host=host_name, dbname=data_base , user=user_name, password=pwd,port=port_id)  
    print("Connection Establised")
    
    # In order to do transaction with database we will need to open a cursor
    # A cursror is something that helps you performt he transaction
    
    # In order to retrieve data from the database using cursor we need to pass a parameter as
    # Ther parameter we have passed is used to fetech data using column name
    cur=conn.cursor(cursor_factory=psycopg2.extras.DictCursor)
   
    
    # Create the select query
    select_script = '''
       select * from emp where salary < 440000
    '''
    
    # Now we will execute the select query string on the emp table
    cur.execute(select_script)
    
    # Now when the query is successfully executed some records will be returned according to the given condition
    # We can use the for loop to get all records and display the data one by one
    # The data is retried in the form of key/value pair (dictinary)
    for record in cur.fetchall():
        print(record['name'],record['dept'], record['designation'],record['salary'])
      
    
except Exception as error:
    print("Exception : " + str(error))
else:
    print("No error generated during execution of Try Block")
finally:
    # It is a good practice to add cursor close and database close statement in the finally block so that these statements are executed always
    if cur is not None:
        cur.close()
        print("Cursor Closed")
        
    if conn is not None:
        conn.close()
        print("Connection Closed")
```
