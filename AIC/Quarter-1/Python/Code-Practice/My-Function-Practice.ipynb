```python
# Python-III Lecture-7 Functions : Dealing with Unknow Number of Arguments
# In Python we can specify a arguments that can handle multiple number of parameters

# For Example : Consider a pizzaOrder Function
def pizzaOrder(size,flavour, toppings):
  print(f"Your pizza of size {size} inches with flavour {flavour} having toppings {toppings}")

pizzaOrder(12,"Fajita", "Olives")
# Now the issue is that there can only be one topping. But in actual we can have multiple toppings
# 

def pizzaOrder(size,flavour,*toppings):
  print(f"Your pizza of size {size} inches with flavour {flavour} having toppings {toppings}")

print(" After Redefining pizzaOrder function with arbitrary number of arguments")
pizzaOrder(12,"Fajita", "Olives")
pizzaOrder(12,"Chicken Tikka", "Chezious","Chicken")
pizzaOrder(12,"Fajita", "Olives","Meat cubes", "Extra Cheeze")
```

```python
# Python-III Lecture-8 Functions : Passing Information Back from them
# Let us create a function that can return a value

def addNumbers(no1,no2):
  ans=no1+no2
  print("The answer is ", ans)
  return ans

a=addNumbers(10,20)
print("The function returned value ", a)

# We can also return more than one values from functions as
print("Example to show example to return multiple values using return statement")
def addNumbers(no1,no2):
  ans=no1+no2
  print("The answer is ", ans)
  return no1,no2,ans

res=addNumbers(10,40)
print("The function returned values ", res)
# Note : The function will return the value in the form of tuple

# We can also return different types of values of functions
print("Example to show example of different types of values using return statement")
def addNumbers(no1,no2):
  ans=no1+no2
  return "The answer is ",ans

res=addNumbers(30,40)
print("The function returned values ", res)
# Note : The function will return the value in the form of tuple

# Moreover we if we write some statments inside the function after the funtion return statements then there will be no error But these 
# statements will never execute
print("Example to show the statements written after return statement")
def addNumbers(no1,no2):
  ans=no1+no2
  return "The answer is ",ans
  print("This message will not be printed")

res=addNumbers(50,45)
```

```python
# Python-III Lecture-9 Functions : Using Functions as Variable (Which is what they really are)
# Functions can be used as variables . This is done by using function call in our expression


def addNumbers(no1,no2):
  return no1+no2

def subtractNumbers(no1,no2):
  return no1-no2

result = addNumbers(50,15) + subtractNumbers(50,15)
# 100          65                35
print(result)

# In that case first of all subtractNumbers(50,15) will be called then will call addNumbers(50,15)
# Then result of both functions will be added and result will be stores in the result variable
```

```python
# Python-III Lecture-10 Functions : Local vs Global Variables
# Local variables are the variables defined inside the funtions. There scope is only inside the function. They are not accessible outside the function
# Global variables are the variables defined outside the function and can be accessed and modified in and outside the functions

def beHappy():
  name = "Mr. Asad"  # Name is a local variable
  print(f"{name} is very happy today")
```

```python
beHappy()
#print(name) # Will generate error as the variable name is not accessible outside the function beHappy()
```

```python
anotherName="Mr. Baber"  # Global Variable

def beSad():
  print(f"{anotherName} is sad today")
```

```python
beSad()
print(anotherName) # antoher Name can be accessed inside and outside of the function 
```

```python
# PIII: Functions within Functions
# We can call a function inside another function providing the required signature of the function




def salaryCalc(basic):
  def commisionCalc(salary):
    if salary >= 80 :
      return salary*80
    elif salary >= 50:
      return salary * 50
    elif salary >= 20:
      return salary*20
    else:
      return 0
  grossSalary = basic + commisionCalc(basic)
  print(f"Your Gross Salary  is = {grossSalary}")
```
