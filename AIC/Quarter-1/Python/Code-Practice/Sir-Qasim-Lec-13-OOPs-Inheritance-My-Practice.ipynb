```python
class Parent():
  def __init__(self):
    self.id=None
    self.name=None
    self.fname=None

  def eat(self,food):
      return f"{self.name} is eating {food}"

  def speak(self,words,language="Urdu"):
      return f"{words}, language = {language}"

class Child(Parent):
    def walk(self,distance):
      return f"{self.name} walked for {distance}"

```

```python
p1=Parent()
c1=Child()

p1.id=1
p1.name="Muhammad Asghar"
p1.fname="Abdul Raheem"

p1.eat("Pizza")

c1=Child()
c1.eat("Biryani")
```

```python
# Multi Class Inheritance ( One to Many)
# That a class is inherited by more than one class
# Consider we have a class Father and Mother
class Father():
  def __init__(self,name):
    self.name=name
    self.fname=None
    self.color='Whistish'
    self.hair_color='Dark Brown'

    def speak(self):
      return "Speaking"

class Mother():
  def __init__(self):
    self.height="5.1 feet"
    self.color="White"

  def silent(self):
    return "Mother is silent"

class Child(Father,Mother):
  pass
# Note  First Child class will be inherited by Mother and then Father
# So the Mother attributes will not be visible as they will be over riddemn

class Child(Mother, Father):
  pass
```

```python
c1=Child("Zeeshan") 
print([i for i in dir(c1) if "__" not in i])
# The above print function will print all methods and attributes of Child Class
```

```python
# Polymorphism
# if object behanve multiple behaviours , with same method name or operator 
# for example + operator works differently with numbers and string

# Overloading  : Multiple behaviours with in same class 
# Before the Python language 
  # def add(int num1,int num2) : return num1+num2
  # def add(float num2,float num2) : return num1+num2
  # def add(int num2, float num2) : return num1+num2
  # def add(float num2, int num2) : return num1+num2
# Method name is same but data types of arguments are different and number of armguments can also very

# In Python we does not worry about variables date types and no of argument
def add(self, *num):
  return sum(num)


# Let us create a Class
class my_Math():
  def __init__(self):
    pass

  def add(self, *num):
    return sum(num)
# In the method add we have given *num which means any number of paramether with any type of values
# the values will be passed as tuple
# But if we use **num then we are informing to use dictionary instead of tuple

ob1=my_Math()
ob1.add(10,20,30.4)
ob1.add(10,20,30,40,50,60)

```

```python
# In Python 
# Overridding : Multiple behaviours with different class
# Consider Example
class Mother():
  def speak(self):
    return "Mother Speak()"

  def write(self):
    return "Mother Write()"

class Child(Mother):
  def speak(self):
    return "Child Speak()"

c1=Child()
c1.speak() # Here child has a Mother Speak Function and Child also have his own speak() .
# So due to the Over Ridding child speak function will be executed

c1.write()
```
