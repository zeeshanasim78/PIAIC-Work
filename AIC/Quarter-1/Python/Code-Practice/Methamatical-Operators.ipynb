```python
## Python Arithmetic Operators
## Arithmetic operators are used with numeric values to perform common mathematical operations:
##    +	Addition  e.g: 	x + y	
##    -	Subtraction	 e.g: x - y	
##    *	Multiplication  e.g: 	x * y	
##    /	Division  e.g: 	x / y	
##    %	Modulus  e.g: 	x % y	
##    **	Exponentiation  e.g: 	x ** y	
##    //	Floor division  e.g: 	x // y	

a = 23
b = 34
c = a + b
print("The value of variable c is = ", c)

value1= 100  ## Integer Value
value2= 10.5  ## Floating Point Type
sum=value1 * value2 ## Result will be in Floating Point
print(sum)
## Note if any variable is of Floating Typpe in Mathematical Expression the result will be in the Floating Point Data Type

value1= 100  ## Integer Value
value2= 10  ## Integer Point Type
sum=value1 / value2 ## Result will be in Floating Point as by defualt division is in Floating Point in Python
print(sum)



##  For division result if any value is Floating Type with // (Floor) operator
## The result will be flaoting with with Floor value
a=100.5
b=6
var=a//b  ## // operator is used instead of / Operator
print(var)  ## The result will be 16.0 instead of 16.66


a="Ali"
b="Azmat"
print(a+ " " + b)

```
