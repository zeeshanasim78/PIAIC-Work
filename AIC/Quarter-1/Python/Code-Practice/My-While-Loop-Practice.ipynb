```python
# PIII : Lecture 12 : While Loops
# There is another type of block called While Loop
# The While loop work simialr to the loops , but the differ in the sence as it allows user to terminate loop by setting flags
# Syntax
# while condition:
#    statements 
a=0
while a<10:
  print(f"Loop Iteration # {a}")
  a+=1
```

```python
# Conside we want take input from user about his/her favourite foods
# Now this can be different for different people. So handle this we use while loop

flag=True;
favFoods=[];

while flag==True:
  userInput=input("Enter you Favourite or press Q to quit");
  if userInput=="Q":
      flag = False
  else:
    favFoods.append(userInput)

# Print Favourite Foods
print(favFoods)
```
