```python
class Country():
    def __init__(self, pop, grow):
        self.pop = pop
        self.grow = grow
    def project_population(self, years):
        growth = self.pop * self.grow
        print(f"Annual popoulation growth is {growth}")
        add_pop = growth * years
        proj_pop = self.pop + add_pop
        print(f"The projected population of the city is {proj_pop} million after {years} years")

class City(Country): 
    def __init__(self,pop, grow):
        Country.__init__(self, pop, grow)
        self.name = None
        self.revenue = None
        
Pakistan = Country(250, .05)
print("\n")
print([i for i in dir(Pakistan) if "__" not in i])

city1 = City(250, .05)
print([i for i in dir(city1) if "__" not in i])
```
