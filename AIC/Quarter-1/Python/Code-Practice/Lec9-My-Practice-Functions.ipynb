```python
# We can also define functions with  unlimited argument
def sum_values(num1,num2,*num):
  #  *num will mean unlimited parameters separated by , and will be stored in a tuple
  # Calulate values using tuple
  # for the time being we will return a test value
  sum_tuple_no=0
  for i in num:
    sum_tuple_no=sum_tuple_no+i
  return num1+num2+sum_tuple_no
```

```python
# We can call the unlimited argument as
a=sum_values(10,20)
print("Sum of two values = ", a)

# We can call the unlimited argument as
a=sum_values(10,20,30,40,50)
print("Sum of all values = ", a)

# We can call the unlimited argument as
a=sum_values(10,20,30,40,50,100,200,300)
print("Sum of all values = ", a)
```

```python
# Generator Function
# Python provides a generator to create your own iterator function. 
# A generator is a special type of function which does not return a single value, instead, it returns an iterator object with a sequence of values.
# In a generator function, a yield statement is used rather than a return statement.

# In simple words we can say Generator function does not do work at a time.
# It will do 1st step and then after call will generate next value


# MY Generator Function
def my_gen_function(start,end):
  for i in range(start,end+1):
    yield i # Yield specifies that this is a generator function

abc = my_gen_function(1,10)
print(abc)
print(next(abc))
print(next(abc))
print(next(abc))
```

```python
# Recursive Function
# Recursion is a common mathematical and programming concept. It means that a function calls itself. This has the benefit of meaning that you can loop through data to reach a result.

def countdown(x):
  print(x)
  if x==0: # Terminate Recursive
    return 0
  else:
    return countdown(x-1)
    
countdown(5)
```
