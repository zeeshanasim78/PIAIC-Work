```python
# Error Handline : Syntax of Try Block
"""
try:
  Write statement where error occurance chances exists
except  (Exception Error Class)
  statement
else:
  statement if no error is generated
finally:
  always run
"""

a=7
b=2
try:
  c=a/b
  print("The answer of a/b is = ",c)
except (ZeroDivisionError):
  print("Zero Division Error")
except(Exception):
   print(" Error Executed")
else:
  print("No Error Generated so Else Block Executed")
finally:
   print("Finally Block will always Executed")


print("Ending Line Out of Try Block")
```

```python
# Code to handle Dynamic Error with Else & Finally block
# Else block is executed with no error is generated

print("Exception Handling Example")
a=7
b=2
l=[1,2,3,5,6,7,8,9,]

try:
  c=a/b
  print("The Result of a/b = ",c)
  print(" The element at the sixth index of list l is ",l[6])
  # open("abc.txt")
except Exception as ex:  # To Dynamically Handle the Error of any type
  print(" Error generated : ", ex)
else:
  print("This is the else block")
finally:
  print("This block will always be executed")

print(" End of the Example Handling ")
```
