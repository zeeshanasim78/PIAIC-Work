```python
# Type Casting
var=int(input("Enter Some Value"))
print(f"The value of var is {var} and its data type is {type(var)}")

ft=float(var)
print(f"After conversion of var value into float the result will be {ft}")

var=int(ft)
print(f"After converting back ft value to int we will get results {var}")
```
