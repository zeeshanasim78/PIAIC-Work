```python
# Lecture # 14 : OOP Encapsulation
# In Encapsulation we can make methods or attributes private
# Private are those methods and attributes that we can access with in a class  or through inheritance.
# It is not accessible outside of the class. If you want to access it outside the class we have to create set and get functions
# To make any attribute private  we use the keyword private
class Facebook():
  def __init__(self,user):
    self.user=user
    self.__password='123'  # if we use __ before a attribute name it will make that attribute private
    self.__email=None

  def __connectDB(self):  ## This is a private method
    return "DB MYSqlSever Connection in Progress"

  def updatePassword(self,newPassword):
    self.__password=newPassword

  def updateEmail(self,email):
    self.__email=email;

  def getEmail(self):
    return self.__email

  def displayInfo(self):
    str1= f"""
    User Name : {self.user}
    Password : {self.__password}
    Email Address : {self.__email}
    """
    str1 = str1 + self.__connectDB()
    print(str1)

  
```

```python
u1=Facebook("Asim") 
print(u1.user) ## This will print user information
#print(u1.password) ## This will generate error as password id declared as private
#print(u1.__password) ## This will generate error as password id declared as private



u1.displayInfo()

u1.updateEmail("zeeshan@gmail.com")
u1.updatePassword("zeePass")
u1.displayInfo()
```

```python
## Abstract CLass
# In OOPs there are classess whose objects we did not create
# We always create its child class Object
# ForExample in Real World there can be a class Mammels whose object we cant create
# Abstract Class can also be accessed in Child Class Body

# In Python 2.7 we create Abstract Class as 
# class A:  Note : No () as we are defining a Abstract Class

# Now from Python 3.7 we have to use a package
from abc import ABC,abstractmethod

 # Abstract Class
class Human(ABC): 
  @abstractmethod
  def __init__(self,name,age):
    self.name=name
    self.age=age

class Male(Human):
  def __init__(self,name,age,height,color):
    super().__init__(name,age)
    self.height=height
    self.color=color


  def displayInfo(self):
    str1 = f"""
    Name = {self.name}
    Age =  {self.age}
    Height = {self.height}
    Complextion = {self.color}
    """
    print(str1)
```

```python
m1=Male("Zeeshan",44,5.7,"brownish")
m1.displayInfo()
```

```python
# File Handling
# Whenever we deal with files like audio, text, image we are doing file handling
# Whenever we open a file we specify some attribute like mode
# Available Modes are r,r+,w,w+,a,a+

with open("abc.txt",'r+') as f:
  a=list(reversed(f.readlines()))
  print(a)
  b="\n The is the additional Text that will be addedd"
  b=b+str(a)
  f.seek(0) # Set Cursor position at start
  f.write(b) # Write at the start  of the file
  f.close() # Close the wile

#Note  : we have use with block so after the block is completet
#It will automatically close the file and release all resoruces

# So it means if we did not use f.close() in with block it will automally close if we havent mentioned it
```
