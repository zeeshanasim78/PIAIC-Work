```python
# Regular Express Example
import re  # To import Regular Express package
text = """
pakistan
"""
# To find all a in the text we can use RE as 
re.findall("a",text)
```

```python
text = """
pakistan America  adad li      222
"""
# To find all ki in the text we can use RE as 
re.findall("ki",text)
```

```python
text = """
pakistan \n America  
pakistan
pakistan 
 adad  222
"""
# To find all \n in the text we can use RE. We use r for escape sequence
re.findall(r"pakistan\n",text)
```

```python
text = """pakistan zindabad
Beautifull country paksitan
pakistan is no 1
we love pakistan
"""
# \b is for boundry . we can use to specify the start and end 
# To find all pakistan word only we use \b for exact word pakistan
# \r must be used for the escape sequence
re.findall(r"\bpakistan\b",text)  # will return ['pakistan', 'pakistan', 'pakistan']

re.findall(r"^pakistan",text) # will return ['pakistan'] means line start with pakistan


```

```python
text = """pakistan zindabad
Beautifull country paksitan
pakistan is no 1
we love pakistan
"""
# 
re.findall(r"^pakistan.*",text) # will return pakistan.* means any character 
# But this will not work for all lines and will stop after 1st line
# to search for all text in all lines we use
re.findall(r"^pakistan.*",text,re.MULTILINE)
```

```python
text = """abc available
aaa abcd 
abcdef  ddabcd
"""
# 
print(re.findall("a*",text)) # will return all that start with a and have any thing after a
```

```python
text = """
Pakistan + America  
pakistanhg
pakistan \\
 adad  222
"""
# \b is for boundry . we can use to specify the start and end 
# To find all pakistan word only we use \b for exact word pakistan
# \r must be used for the escape sequence
# re.findall("+",text) # This will generate error as + is a special character
# to find special character we will use \
re.findall(r"\+",text) # To find + in the string

# Similarly to search \ we will use 
re.findall(r"\\",text)
```

```python
v1="abc as ad abbb baa" # *,+
re.findall("ab?",v1) # will return ['ab', 'a', 'a', 'ab', 'a', 'a']
re.findall("ab?|abb?",v1) # will return ['ab', 'a', 'a', 'ab', 'a', 'a']
```

```python
x="""
name: Zeeshan Asim
father's name: M. Ashgar
education: MSDS
age:30
cnic:342345345345
"""

re.findall(r"age:(\d+)",x)  # want to extract 30
# \d means number + repeatation
# will return age value

# to find cnic we can use RE as
re.findall(r"cnic:([0-9]+)",x)  # means all numbers betwen 0 to 9 after cnic  , +  will allow any number of nos
```

```python
txt="""
name:zeeshan
age:40
cnic=343432342342
email:abc@gmail.com"""
re.findall(r"([0-9]+)",txt) # to find all numbers

```

```python
txt="""
asdasd aslkdjha laksdjlakdsja
hbc@ubl.com asjdkhakjd akjsdh test@hotmail.com
asdasd ajksdh kjasdh akjsdhkajsdh indo@gmail.com
email:abc@gmail.com
abc.123@gmail.com"""
 
# to find eamil addresses in a text string

# to find hbc@gmail.com
re.findall(r"hbc@ubl.com",txt) # will return ['hbc@ubl.com']

# to find all email address that have doamin gmail.com
# \w means to find word
re.findall(r"\w+@gmail.com",txt) # will return ['indo@gmail.com', 'abc@gmail.com']

# to find all emaill addresses that end with.com we can use regular express as 
re.findall(r"\w+@\w+.com",txt) # will return ['hbc@ubl.com', 'test@hotmail.com', 'indo@gmail.com', 'abc@gmail.com']

# to find all email address that end with any name
re.findall(r"\w+@\w+.\w+",txt) 
# will return ['hbc@ubl.com', 'test@hotmail.com',  'indo@gmail.com',  'abc@gmail.com',  '123@gmail.com'] 
# But it will not return . email addresses 

re.findall(r"[0-9a-zA-Z.]+@.[0-9a-zA-Z.]+.com",txt)
# will return ['hbc@ubl.com',  'test@hotmail.com',  'indo@gmail.com',  'abc@gmail.com',  'abc.123@gmail.com'] 

```
