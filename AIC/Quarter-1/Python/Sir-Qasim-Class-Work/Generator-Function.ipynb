```python
def square_numbers(nums):
  result = []

  for i in nums:
    result.append(i**2)
  return result

my_nums = square_numbers([1,2,3,4,5])

print(my_nums)
```

```python
# Lets create a Generator Function who can do this work in a more efficient way
# Yield will return the value one by one

def square_numbers(nums):
  for i in nums:
    yield(i**2)

my_nums = square_numbers([1,2,3,4,5])
print(my_nums)
  

```

```python
# Now to get the values we will use it as
next(my_nums)
# This will return one value at a time
# On Every next it will generate the next value
```
